```python
import pickle as pickle
from collections import OrderedDict
from matplotlib import pyplot as plt

models = ["knn", "user_cluster", "item_cluster", "overall"]
min_users = 10

# plot
for model in models:
    qq = pickle.load(open( "ndcgs/{}.pkl".format(model), 'rb'))
    ndcgs = qq["ndcgs"]
    mean_ndcgs = OrderedDict( { k: sum( v ) / len( v ) for k, v in ndcgs.items() if len( v ) >= min_users } )
    for k, v in list(mean_ndcgs.items())[:20] :
        pass
        #print(k, v)
    plt.plot( list(mean_ndcgs.keys()), list(mean_ndcgs.values()))

axes = plt.gca()
axes.set_ylim([0,0.02])

plt.show()
```
